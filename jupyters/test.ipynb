```python
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import Docx2txtLoader
```

```python
loader = Docx2txtLoader("./docs/Dados HUBot - Empreendedorismo.docx")
data1 = loader.load()
data1
```

```python
loader = Docx2txtLoader("./docs/Dados HUBot - Institucional.docx")
data2 = loader.load()
data2
```

```python
loader = Docx2txtLoader("./docs/Dados HUBot - Women In Tech.docx")
data3 = loader.load()
data3
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n\n\n",
    chunk_size=6000,
    chunk_overlap=400,
    length_function=len,
    is_separator_regex=False,
)
chunks1 = text_splitter.split_documents(data1)
chunks1
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n\n\n",
    chunk_size=13000,
    chunk_overlap=400,
    length_function=len,
    is_separator_regex=False,
)
chunks2 = text_splitter.split_documents(data2)
chunks2
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n\n\n",
    chunk_size=8000,
    chunk_overlap=400,
    length_function=len,
    is_separator_regex=False,
)
chunks3 = text_splitter.split_documents(data3)
chunks3
```

```python
chunks = [chunks1, chunks2, chunks3]
all_chunks = []
for chunk in chunks:
    for document in chunk:
        all_chunks.append(document)
```

```python
all_chunks
```

```python
from dotenv import load_dotenv
import os
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
load_dotenv()


embeddings = OpenAIEmbeddings(api_key=OPENAI_API_KEY, 
                              model="text-embedding-3-large")
vectorstore = FAISS.from_documents(all_chunks, embeddings)
```
